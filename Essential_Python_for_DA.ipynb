## **OOP**

* attributes: name, species, age
* methods: function bark() hbd()

```python
class Dog:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self.age = age
    def bark(self):
        print(f"{self.name} is barking 'hong hong'!")

    def hbd(self):
        self.age += 1
```

```python
## attributes
dog1 = Dog("milo", "Chihuahua",7)
dog2 = Dog("david", "Husky",5)
```

```python
print(f"{dog1.name} is {dog1.species}")
```

```python
print(f"{dog2.name} is {dog2.species}")
```

```python
dog1.bark()
```

```python
dog2.bark()
```

```python
print(dog1.age)
dog1.hbd()
print(dog1.age)
```

```python
## homework OOP for an ATM

## deposit, withdraw, check balance, OTP, order coffe, pay bill, internet
```

```python
class ATM:
    def __init__(self, username, bank, amount):
        self.username= username
        self.bank = bank
        self.amount = amount

    def __str__(self):
        return f"{self.username} uses {self.bank} and has {self.amount} THB in an account."

    def check_balance(self):
        print(f"Balance: {self.amount} THB")

    def deposit(self, saving):
        self.amount += saving
        print(f"Deposit {saving} THB successfully!")
```

```python
gern = ATM("siri gern", "kbank", 1000)
```

```python
gern.check_balance()
```

```python
gern.deposit(300)
```

```python
gern.check_balance()
```

```python
print(gern)
```

### **Super Class**

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

##extension
class Employee(Person):
    def __init__(self, name, age, company):
        super().__init__(name, age)
        self.company = company

    def greeting(self):
        print("Hello World!")
```

```python
gern = Employee("gern", 21,"MomDad")
```

```python
gern.name
```

```python
toy = Person("toy",37)
```

```python
gern.greeting()
```
