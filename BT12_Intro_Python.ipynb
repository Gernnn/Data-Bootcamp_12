## **Control Flow**

* if
* for
* while

````python
# if
score = 82
if score >= 80:
    print("passed")
else:
    print("failed")
````

````python
# mulitple grades
score = 44

if score >= 80:
    print("A")
elif score >= 70:
    print("B")
elif score >= 60:
    print("C")
else:
    print("D")
````

````python
# get input from user
username = input("What's your name: ")
age = int(input("how old are you: "))
````

````python
# multiple conditions
# night club
has_id = True
age_over_18 = False
vip_customer = False

if has_id and (age_over_18 or vip_customer):
    print("Welcome to the club.")
else:
    print("Please come next time.")
````

````python
# for & while
items = ["egg","milk","bread"]

for item in items:
    if item == "milk":
        print("CP Meiji")
    else:
        print(item)
````

````python
## list comprehension
result = [ ]

for item in items:
    result.append(item.upper())

print(result)
````

````python
# list comprehension
[item.upper() for item in items]
````

````python
# while
while True:
    user_input = input("Play on: Y/N ")
    if user_input == "N":
        print("Goodbye!")
        break
    else:
        print("Gooooo!")
````

````python
## while another example
count = 0

while count < 5:
    print("hi!")
    count += 1
````

## **Object Oriented Programming/ OPP**



````python
class Dog:
    name = "Gern"
    age = 5
    color = "Brown"
    breed = "French Bulldog"

 # function (Dog metod)
    def sitting(self):
        print("I am sitting now!")
    def hungry(self, food_name):
        print(f"I am hungry, I need {food_name}!")
````

````python
my_dog = Dog()
type(my_dog)
````

````python
my_dog.hungry("hotdog")
````
